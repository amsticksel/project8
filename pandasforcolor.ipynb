```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib tk
```

```python
df = pd.read_csv('dogs_clean.csv')
```

```python
import json
primary = []
secondary = []
tertiary = []
for k,v in df.iterrows():    
    c = eval(v.colors)       
    primary.append(c['primary'])
    secondary.append(c['secondary'])
    tertiary.append(c['tertiary'])
    

df['primary'] = primary
df['secondary'] = secondary
df['tertiary'] = tertiary

df['primary'] = df.primary.str.split('/')
df['secondary'] = df.secondary.str.split('/')
df['tertiary'] = df.tertiary.str.split('/')
```

```python
df['status'] = df.status.replace({"adoptable":"not_adopted"})
```

```python
df['primary'] = df.primary.str.join(sep='')
```
