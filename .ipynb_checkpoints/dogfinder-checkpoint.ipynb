```python
#####pulling up the csv file to look at it
dogsdf = pd.read_csv('dogs.csv')
dogsdf
```

## Anne's Stuff - Looking at dog ages

```python
####Creating a dataframe with the status and age of the dogs
agesdf = pd.crosstab(dogsdf.status, dogsdf.age)
agesdf
sumage = agesdf.sum(axis=1)
agesdf['Total'] = sumage
agesdf.drop('found')
```

```python
##Calculating the percentage of dogs in each group
baby = (agesdf['Baby']/agesdf['Total'])*100
young = (agesdf['Young']/agesdf['Total'])*100
adult = (agesdf['Adult']/agesdf['Total'])*100
senior = (agesdf['Senior']/agesdf['Total'])*100
baby, young, adult, senior
```

```python
####creating a dataframe and a graph to group the dogs by status
dogdatageams = {'Baby': baby,
                'Young': young,
                'Adult': adult,
                'Senior': senior}
index = ['baby', 'young', 'adult', 'senior']
adoptable = [32.36, 29.32, 34.31, 4.01]
adopted = [26.49, 29.04, 39.48, 5.00]

dfdogagesams = pd.DataFrame({'Adopted': adopted, 'Adoptable': adoptable}, index=index)
ax = dfdogagesams.plot.bar(rot=0)
plt.title('Dogs By Age Group')
plt.xlabel('Age Group')
plt.ylabel('% of dogs in group')
plt.savefig("dogages.png")
```

## Anne's Stuff - Looking at friendliness

```python
#####further cleaning the data to fill in missing info
dfams = dogsdf.loc[:,"spayed_neutered":].fillna('missing')
dfams
df2ams = dogsdf.loc[:,:"spayed_neutered"]
df3ams = pd.concat([df2ams,dfams],axis=1)
df3ams

#creating dataframe columns for each type of friendliness
df3ams = pd.get_dummies(df3ams, columns=['friendly_cats','friendly_dogs','friendly_children'])
means = df3ams.groupby('status').mean()
counts = df3ams.groupby('status').count()
stds = df3ams.groupby('status').std()
df3ams

sum1 = df3ams['friendly_children_True'].sum()
sum1

sum2 = df3ams['friendly_children_False'].sum()
sum2


sum3 = sum1 + sum2
sum3
```

```python
#calculating the means, counts, and standard deviation for each one
counts = df3ams.groupby('status').count()
means = df3ams.groupby('status').mean()
stds = df3ams.groupby('status').std()
```

```python
counts, means, stds
```

```python
means.drop('found')
stds.drop('found')

fig, axes = plt.subplots(1,3, figsize=(10,5))
fig.suptitle('Dog Friendliness')
ylim = (-.2, 1)


plt.setp(axes, ylim=ylim)

axes[0].set_title('Cats')
axes[0].text(-.5, .9, 'pvalue=1.3754603842630517e-20',bbox=dict(facecolor='red', alpha=0.5))
means['friendly_cats_True'].drop('found').plot.bar(yerr=stds['friendly_cats_True'],ax=axes[0])

axes[1].set_title('Other Dogs')
axes[1].text(-.5, .9, 'pvalue=3.751010370441551e-73', bbox=dict(facecolor='red', alpha=0.5))
means['friendly_dogs_True'].drop('found').plot.bar(yerr=stds['friendly_dogs_True'],ax=axes[1])

axes[2].set_title('Children')
axes[2].text(-.5, .9, 'pvalue=6.909262892411688e-49', bbox=dict(facecolor='red', alpha=0.5))
means['friendly_children_True'].drop('found').plot.bar(yerr=stds['friendly_children_True'],ax=axes[2])

plt.savefig('dogfriendly.png')
plt.show()
```
